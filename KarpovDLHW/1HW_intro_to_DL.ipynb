```python
import torch, torch.nn as nn
```

```python
tens = torch.randn(3, 5)
```

```python
tens
```

```python
def function1(tensor: torch.Tensor, count_over: str):
    if count_over == 'columns':
        return torch.sum(tensor, dim=1)
    if count_over == 'row':
        return torch.sum(tensor, dim=0)
function1(tens, 'row'), function1(tens, 'columns')
```

```python

def function2():
    w = torch.randn(2, dtype=torch.float32, requires_grad=True)
    return w
    
    
```

```python
n_features = 2
n_objects = 300

w_true = torch.randn(n_features)
X = (torch.rand(n_objects, n_features) - 0.5) * 5
Y = X @ w_true + torch.randn(n_objects) / 2
```

```python
def function3(X: torch.Tensor, Y: torch.Tensor):
    w = function2()
    step = 1e-2
    n_steps = 200  
    for i in range(n_steps):
        y_pred = X @ w
        mse = torch.mean((Y - y_pred)**2)
        print(f'MSE на шаге {i + 1} {mse.item():.5f}')
        mse.backward()
        with torch.no_grad():
            w -= w.grad * step
        w.grad.zero_()
    return w
function3(X, Y)

        
    
    
```

```python
def function4(X: torch.Tensor, Y: torch.Tensor):
    w = function2()
    step = 1e-2
    n_steps = 200
    layer = nn.Linear(in_features=2, out_features=1)
    for i in range(n_steps):
        y_pred = layer(X).ravel()
        mse = torch.mean((Y - y_pred)**2)
        print(f'MSE на шаге {i + 1} {mse.item():.5f}')
        mse.backward()
        with torch.no_grad():
            layer.weight -= layer.weight.grad * step
            layer.bias -= layer.bias.grad * step
        layer.zero_grad()
    return layer
function4(X, Y)
```
