```python
df = df.drop('hash', axis = 1)
```

```python
df.head()
```

```python
df_sorted = df.sort_values(by=['Учебная группа', 'Номер ЛД', 'Учебный год', 'Полугодие'])
```

```python
df_sorted.head(60)
```

```python
df_sorted['Год начала обучения'] = df_sorted['Учебная группа'].str.extract(r'(\d{2})').astype(int) + 2000

# Преобразование учебного года и полугодия в порядковые значения
df_sorted['Год начала семестра'] = df_sorted['Учебный год'].str.split(' - ').str[0].astype(int)

# Функция для вычисления номера семестра
def calculate_semester(row):
    year_difference = row['Год начала семестра'] - row['Год начала обучения']
    if 'II' in row['Полугодие']:
        semester_number = year_difference * 2 + 2
    elif 'I' in row['Полугодие']:
        semester_number = year_difference * 2 + 1
    return semester_number

# Вычисление номера семестра
df_sorted['Номер семестра'] = df_sorted.apply(calculate_semester, axis=1)
```
